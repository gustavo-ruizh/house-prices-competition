### 1.2 Set up visualization tools

```python
# Create function that plots categorical distribution
def plot_cat_(df, col, ax, hue=None):
    """Function that plots countplot of categorical column 
    from given df on given ax"""
    _ = sns.countplot(y=col, data=df, ax=ax, \
                      palette='Blues_r', hue=hue) 
    _ = ax.tick_params(labelsize=12)
    _ = ax.set(xlabel=None, ylabel=None)
    _ = ax.set_title(col+' count', fontsize=14)
    _ = ax.set_xlim(0, df[col].value_counts().max()*1.2)
```

```python
# Create function that plots boxplot
def plot_box_(df, x, y, ax):
    """Function that plots a boxplot
    for given dataframe columns x and y
    on given ax"""
    _ = sns.boxplot(y=y, x=x, \
                    data=df, ax=ax, \
                    palette='Blues_r', \
                    showfliers=False)
    _ = ax.set_xlabel(x, fontsize=12)
    _ = ax.set_ylabel('', fontsize=12)
    _ = ax.set_title(y, fontsize=12)
    _ = ax.tick_params(labelsize=10)
```

```python
# Create function that plots KDE
def plot_kde_(name, column, ax):
    """Function that plots the kernel density
    estimate (KDE) for given dataframe column
    on given ax"""
    (mu, sigma) = norm.fit(column)
    skewness = column.skew()
    kurtosis = column.kurt()

    sns.kdeplot(column, ax=ax)
    ax.legend(['mean= {:.0f}\n'.format(mu)+\
               'std= {:.0f}\n'.format(sigma)+\
               'skewness= {:.2f}\n'.format(skewness)+\
               'kurtosis= {:.2f}'.format(kurtosis)], \
              loc='best', fontsize=12, frameon=True)
    ax.set_ylabel('Density', fontsize=14)
    ax.set_xlabel(name, fontsize=14)
    ax.tick_params(labelsize=12)
```

```python
# Create function that plots ECDF
def plot_ecdf_(name, column, ax):
    """Function that plots the empirical cumulative
    distribution function (ECDF) for given dataframe
    column on given ax"""    
    x_cdf = np.sort(column)
    y_cdf = np.arange(1, len(x_cdf) + 1) / len(x_cdf)
    
    x_norm = np.linspace(np.min(x_cdf), np.max(x_cdf))
    y_norm = norm(np.mean(x_cdf), np.std(x_cdf)).cdf(x_norm)
    
    ax.plot(x_cdf, y_cdf, marker = '.', markersize = 4, \
            linestyle = 'none', alpha=0.7)
    ax.plot(x_norm, y_norm, color = 'gray')
    
    ax.axvline(x = np.mean(x_cdf), color = 'k', linestyle = '--')
    ax.text(np.mean(x_cdf) * 1.035 , np.min(y_cdf), \
            'mean = ' + str(int(np.mean(x_cdf))), \
            verticalalignment = 'bottom', horizontalalignment = 'left', \
            fontsize = 12)
    
    ax.set_xlabel(name, fontsize=14)
    ax.set_ylabel('ECDF', fontsize=14)
    ax.tick_params(labelsize=12)
```

```python
# Create function that plots scatterplot
def plot_scatter_(x, x_name, y, y_name, ax):
    """Function that plots a scatterplot
    between given dataframe columns x and y
    on given ax"""
    
    # Plot scatterplot
    ax.plot(x, y, marker = '.', markersize = 5, \
            linestyle = 'none', alpha = 0.8)
    ax.set_xlabel(x_name, fontsize=12)
    ax.set_ylabel(y_name, fontsize=12)
    
    # Plot regression line
    res = linregress(x, y)
    rx = np.array([x.min(), x.max()])
    ry = res.intercept + res.slope * rx
    ax.plot(rx, ry, '-', color = 'darkorange')
    
    # Calculate Pearson r coefficient
    r, p = stats.pearsonr(x, y)
    ax.set_ylim(ymax = y.max())
    ax.legend(['r ={:.2f}\np ={:.2f}'.format(r,p)],\
              loc=1, fontsize=12)    
```

### 1.3 Set up validation tools

```python
# Create function that estimates RMSE
def rmse_(y_val,y_pred):
    y_val = [max(v,0) for v in y_val]
    y_pred = [max(p,0) for p in y_pred]
    return np.sqrt(mean_squared_error(np.log1p(y_val), \
                                        np.log1p(y_pred)))
```

```python
# Create function that validates given model
def validate_(model, X, y):
    """Function that evaluates given model
    using 5-fold cross-validation of training data
    and returns RMSE scores"""
    # Create empty list of scores
    cv_scores = []
    # Create 5-fold splitter
    kf = KFold(shuffle=True, random_state=0, n_splits=5)
    # Loop over folds and obtain accuracy score
    for train_ind, valid_ind in kf.split(X):
        X_train, X_valid = X.iloc[train_ind], X.iloc[valid_ind]
        y_train, y_valid = y.iloc[train_ind], y.iloc[valid_ind]
        # Train and assess model
        model.fit(X_train, y_train)
        y_pred = model.predict(X_valid)
        cv_scores.append(np.round(rmse_(y_valid, y_pred), 4))
    
    return cv_scores
```

```python
# Create function to plot validation curve
def plot_val_curve_(name, model, X, y, param_name, param_vals):
    """Function that evaluates given model
    using 5-fold cross-validation of training data
    and plots resulting validation curve"""
    # Get training and validation scores
    train_scores, valid_scores = validation_curve(model, X, y, \
                                 param_name=param_name, \
                                 param_range=param_vals, \
                                 scoring='neg_mean_squared_log_error', \
                                 n_jobs=-1, cv=5)
    # Calculate means and standard deviations
    train_scores_mean = np.mean(np.sqrt(-1*train_scores), axis=1)
    train_scores_std = np.std(np.sqrt(-1*train_scores), axis=1)
    valid_scores_mean = np.mean(np.sqrt(-1*valid_scores), axis=1)
    valid_scores_std = np.std(np.sqrt(-1*valid_scores), axis=1)
    lw=2
    # Plot training curve and 1 SD interval
    plt.plot(param_vals, train_scores_mean, label='training_score', \
            color='steelblue', lw=lw)
    plt.fill_between(param_vals, train_scores_mean-train_scores_std, \
                    train_scores_mean+train_scores_std, alpha=0.2, \
                    color='steelblue', lw=lw)
    # Plot validation curve and 1 SD interval
    plt.plot(param_vals, valid_scores_mean, label='cv_score', \
            color='darkorange', lw=lw)
    plt.fill_between(param_vals, valid_scores_mean-valid_scores_std, \
                    valid_scores_mean+valid_scores_std, alpha=0.2, \
                    color='darkorange', lw=lw)
    # Add labels to graph
    plt.legend(loc='best', frameon=True)
    plt.title('Validation curve with '+name, fontsize=18)
    plt.xlabel(param_name, fontsize=16)
    plt.ylabel('RMSE', fontsize=16)
    plt.tick_params(labelsize=14)
    plt.tight_layout()
```

```python
# Create function to plot learning curves
def plot_learning_curves(name, model, X, y):
    """Function that evaluates given model
    using training data and plots resulting 
    learning curve"""
    X_train, X_val, y_train, y_val = train_test_split(X, y, \
                                     test_size=0.2, random_state=10)
    train_errors, val_errors = [], []
    for m in range(1, len(X_train) + 1, 50):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(rmse_(y_train[:m], y_train_predict))
        val_errors.append(rmse_(y_val, y_val_predict))
        
    plt.plot(range(1, len(X_train) + 1, 50), train_errors, \
             color='steelblue', linewidth=2, label='train')
    plt.plot(range(1, len(X_train) + 1, 50), val_errors, \
             color='darkorange', linewidth=2, label='val')
    plt.hlines(np.mean([train_errors[-1], val_errors[-1]]), \
               0, len(X_train), color='gray', linestyle='dashed', \
               label='mean')
    
    plt.legend(loc='upper right', fontsize=14)
    plt.title('Learning curve with '+name, fontsize=18)
    plt.xlabel('Training set size', fontsize=16)
    plt.ylabel('RMSE', fontsize=16)
    plt.tight_layout()
```

### 7.2 Tune hyperparameters 

```python
# Set random seed
random_state = 75
```

```python
# Ridge regressor
rg = Ridge(random_state=random_state)

rg_params = {'alpha':np.linspace(1, 1500)}

gs_rg = GridSearchCV(rg, param_grid=rg_params, cv=5, \
                     scoring='neg_mean_squared_log_error', \
                     n_jobs=-1, verbose=0)

_ = gs_rg.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_rg.best_params_
rg_best = gs_rg.best_estimator_
```

```python
# Lasso regressor
ls = Lasso(random_state=random_state, max_iter=10e5,)

ls_params = {'alpha':np.linspace(1, 2500)}

gs_ls = GridSearchCV(ls, param_grid=ls_params, cv=5, \
                     scoring='neg_mean_squared_log_error', \
                     n_jobs=-1, verbose=0)

_ = gs_ls.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_ls.best_params_
ls_best = gs_ls.best_estimator_
```

```python
# SVM regressor
svr = SVR()

svr_params = {'C':np.linspace(1e5, 1e6)}

gs_svr = GridSearchCV(svr, param_grid=svr_params, cv=5, \
                      scoring='neg_mean_squared_log_error', \
                      n_jobs=-1, verbose=0)

_ = gs_svr.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_svr.best_params_
svr_best = gs_svr.best_estimator_
```

```python
# Gradient Boosting regressor
gb = GradientBoostingRegressor(random_state=random_state)

gb_params = {'max_depth': [4],
             'learning_rate': [0.09, 0.08, 0.07],
             'n_estimators': [100, 300, 500],
             'max_features': ['sqrt']}

gs_gb = GridSearchCV(gb, param_grid=gb_params, cv=5, \
                     scoring='neg_mean_squared_log_error', \
                     n_jobs=-1, verbose=0)

_ = gs_gb.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_gb.best_params_
gb_best = gs_gb.best_estimator_
```

```python
# Random Forest regressor
rf = RandomForestRegressor(random_state=random_state)

rf_params = {'max_depth': [13, 14, 15],
             'n_estimators': [500, 700, 900],
             'min_samples_split': [2,4,6],
             'max_features': ['sqrt'], 
             'n_jobs':[-1]}

gs_rf = GridSearchCV(rf, param_grid=rf_params, cv=5, \
                     scoring='neg_mean_squared_log_error', \
                     n_jobs=-1, verbose=0)

_ = gs_rf.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_rf.best_params_
rf_best = gs_rf.best_estimator_
```

```python
# XGBoost regressor
xgb = XGBRegressor(random_state=random_state)

xgb_params = {'verbosity':[0], 
              'max_depth': [2],
              'eta': [0.06, 0.07, 0.08],
              'n_estimators': [700, 900, 1100],
              'subsample': [0.5, 0.7, 0.9]}

gs_xgb = GridSearchCV(xgb, param_grid=xgb_params, cv=5, \
                      scoring='neg_mean_squared_log_error', \
                      n_jobs=-1, verbose=0)

_ = gs_xgb.fit(x_train_e_s, y_train)
print('Best parameters:')
gs_xgb.best_params_
xgb_best = gs_xgb.best_estimator_
```
